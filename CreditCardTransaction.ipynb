### Data Cleansing

```python
dataset.drop(columns=['Avg_Open_To_Buy','Months_on_book','Total_Trans_Ct','CLIENTNUM'], inplace=True)
```

```python
Attrition_Flag=pd.get_dummies(dataset['Attrition_Flag'], drop_first=True)
Gender=pd.get_dummies(dataset['Gender'], drop_first=True)
Marital_Status=pd.get_dummies(dataset['Marital_Status'], drop_first=True)
Income_Category=pd.get_dummies(dataset['Income_Category'], drop_first=True)
Card_Category=pd.get_dummies(dataset['Card_Category'], drop_first=True)
```

```python
education_level = {'Unknown':0, 'Uneducated':1, 'High School':2, 'College':3, 'Graduate':4, 'Post-Graduate':5, 'Doctorate':6}
dataset['Education_Level']=dataset['Education_Level'].map(education_level)
```

```python
dataset=pd.concat([dataset,Attrition_Flag,Gender,Marital_Status,Income_Category,Card_Category], axis=1)
```

```python
dataset.drop(columns=['Attrition_Flag','Gender','Marital_Status','Income_Category','Card_Category'], inplace=True)
```

### Data Modeling

```python
X = dataset.drop(['Credit_Limit'], axis=1)
y = dataset['Credit_Limit']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
reg_lr = LinearRegression()
reg_lr.fit(X_train, y_train)
```

```python
y_lr = reg_lr.predict(X_test)
```

```python
reg_lr.score(X_train, y_train)
```

```python
reg_lr.score(X_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print(mean_absolute_error(y_test, y_lr))
print(mean_squared_error(y_test, y_lr))
print(np.sqrt(mean_squared_error(y_test, y_lr)))

print(r2_score(y_test, y_lr))
```

```python
plt.scatter(y_lr, y_test, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_lr")
plt.show()
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(X_train, y_train)
```

```python
y_pred = reg_rf.predict(X_test)
```

```python
reg_rf.score(X_train, y_train)
```

```python
reg_rf.score(X_test, y_test)
```

```python
plt.scatter(y_test, y_pred, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
print(mean_absolute_error(y_test, y_pred))
print(mean_squared_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))

print(r2_score(y_test, y_pred))
```

#### Hyperparameter Tuning

```python
n_estimators = [10,50,70,100,120]
max_features = ['auto', 'sqrt']
max_depth = [10,15,20,25,30]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_params={'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn import metrics
```

```python
random_rfr=RandomizedSearchCV(estimator=reg_rf, param_distributions=random_params,n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
random_rfr.fit(X_train,y_train)
```

```python
random_rfr.best_params_
```

```python
prediction = random_rfr.predict(X_test)
```

```python
plt.scatter(y_test, prediction, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("prediction")
plt.show()
```

```python
metrics.r2_score(y_test, prediction)
```

```python
random_rfr.score(X_test, y_test)
```
